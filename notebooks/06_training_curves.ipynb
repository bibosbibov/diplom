# Кривые обучения mBERT (TensorBoard → PNG)

Парсит все event-файлы из `logs/tensorboard/`, агрегирует эпошные скаляры
(`epoch_train_loss`, `epoch_val_loss`, `macro_f1`) по этапам Stage 1 / Stage 2
и сохраняет график в `reports/figures/training_curves.png`.

```python
import sys
from pathlib import Path

ROOT = Path('.').resolve()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.evaluation import parse_tensorboard_logs, plot_training_curves

scalars = parse_tensorboard_logs(ROOT / 'logs/tensorboard')
for stage, data in scalars.items():
    print(f'{stage}:')
    for key, points in data.items():
        first = points[0] if points else None
        last  = points[-1] if points else None
        print(f'  {key}: {len(points)} точек, первая={first}, последняя={last}')
```

```python
out_png = ROOT / 'reports/figures/training_curves.png'
out_png.parent.mkdir(parents=True, exist_ok=True)
plot_training_curves(scalars, out_png)
print('Сохранено:', out_png)
```
